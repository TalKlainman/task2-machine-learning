## Part 3 - Experiments
You could add as many code cells as needed

```python
# Define features and target variable
X = train_data.drop(columns=['Survived'])
y = train_data['Survived']

# Numeric and categorical features
numeric_features = ['Age', 'Fare', 'SibSp', 'Parch'] 
categorical_features = ['Pclass', 'Sex', 'Embarked']

# Preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Handling missing values
    ('scaler', StandardScaler())])  # Scaling numeric features

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])  # One-hot encoding for categorical features

# Combine preprocessing steps for numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.metrics import precision_score, recall_score, f1_score, make_scorer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import pandas as pd

# Define models
models = {
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

# Define hyperparameters for grid search
param_grids = {
    'K-Nearest Neighbors': {
        'classifier__n_neighbors': [3, 5, 7, 9],  # Number of neighbors to use
        'classifier__p': [1, 2]  # Power parameter for the Minkowski distance (1=Manhattan, 2=Euclidean)
    },
    'Naive Bayes': {
        'classifier__var_smoothing': [1e-9, 1e-8, 1e-7],  # Example values for Naive Bayes
        'classifier__priors': [None, [0.4, 0.6]]  # Example values for priors
    }
}

# Define the scalers to compare
scalers = {
    'Standard Scaler': StandardScaler(),
    'MinMax Scaler': MinMaxScaler()
}

# Custom scoring metrics
scorers = {
    'precision': make_scorer(precision_score, average='macro'),
    'recall': make_scorer(recall_score, average='macro'),
    'f1': make_scorer(f1_score, average='macro')
}

# Storing the results
results = []

# Iterate through each scaler and model
for scaler_name, scaler in scalers.items():
    for name, model in models.items():
        # Create a pipeline with the scaler and classifier
        pipeline = Pipeline(steps=[('scaler', scaler), ('classifier', model)])

        # Define Grid Search
        grid_search = GridSearchCV(pipeline, param_grids.get(name, {}), cv=5, scoring=scorers, refit='f1', n_jobs=-1)

        # Fit the grid search
        grid_search.fit(X, y)

        # Cross-validate to get the scores for Precision, Recall, and F1
        cv_results = cross_validate(grid_search.best_estimator_, X, y, cv=5, scoring=scorers)

        # Store different combinations of hyperparameters and their corresponding metrics
        for i in range(len(grid_search.cv_results_['params'])):
            # Append results for each hyperparameter combination and scaler
            results.append({
                'Method': name,
                'Hyper Parameter': grid_search.cv_results_['params'][i],
                'Scaling Method': scaler_name,
                'Precision': grid_search.cv_results_['mean_test_precision'][i],
                'Recall': grid_search.cv_results_['mean_test_recall'][i],
                'F1': grid_search.cv_results_['mean_test_f1'][i]
            })

```

```python
# Convert results to a DataFrame
results_df = pd.DataFrame(results)

pd.set_option('display.max_colwidth', None)

# Sort the DataFrame by F1 score in descending order
sorted_results_df = results_df.sort_values(by='F1', ascending=False)

# Display the sorted table
print("\nSorted Model Comparison (by F1 Score):")
display(sorted_results_df)
```

## Part 4 - Training 
Use the best combination of feature engineering, model (algorithm and hyperparameters) from the experiment part (part 3)

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import joblib

# Numeric and categorical features
numeric_features = ['Age', 'Fare', 'SibSp', 'Parch']
categorical_features = ['Pclass', 'Sex', 'Embarked']

# Preprocessing for numeric data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine preprocessing steps for both numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
# Build the final model pipeline with the best combination
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', KNeighborsClassifier(n_neighbors=3, p=2))])

# Train the model on the entire training data
pipeline.fit(X, y)

print("Model training completed with the best combination!")
```

## Part 5 - Apply on test and show model performance estimation

```python
y_pred = pipeline.predict(X)
print(y_pred[:5])
print(y[:5].values)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate evaluation metrics
precision = precision_score(y, y_pred, average='macro')
recall = recall_score(y, y_pred, average='macro')
f1 = f1_score(y, y_pred, average='macro')

# Print the results
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
